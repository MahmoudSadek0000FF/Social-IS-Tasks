```python
#exercise 1 answer:
#الحل 
def get_leaves(G):
    nodes_list = ''
    for i in G.nodes:
        if G.degree(i)==1:
            nodes_list += i
```

```python
G = nx.Graph()
G.add_edges_from([
        ('a', 'b'),
        ('a', 'd'),
        ('c', 'd'),
    ])
```

```python
print(get_leaves(G))
```
